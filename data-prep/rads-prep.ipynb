#### Fixing the Time Column. 
##### I decided to set all the nulls to be midnight

```python
rads_raw['Time'].fillna(0, inplace = True)
```

```python
rads_raw.Time.info()
```

```python
rads_raw['Time'] = rads_raw['Time'].astype('int')
rads_raw.Time.dtypes
```

```python
rads_raw['Time'] = rads_raw['Time'].astype('str')
rads_raw.Time.dtypes
```

```python
rads_raw.Time.dtypes
```

```python
# trying to make the times be 4 digits because it doesn't like "0.0" for midnight
rads_raw['Time'] = rads_raw['Time'].apply(lambda x: x.zfill(4))
# turning it to time format
rads_raw['Time'] = pd.to_datetime(rads_raw['Time'], format='%H%M').dt.time

# https://stackoverflow.com/questions/54313461/pandas-convert-float-to-proper-datetime-or-time-object
```
